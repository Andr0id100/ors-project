```python
import matplotlib.pyplot as plt
from pathlib import Path
import random
from random import randint
from PIL import Image, ImageDraw
```

```python
path = Path("content/data/patches_256_v2/")
```

```python
image_paths = list(path.iterdir())
```

```python
fig, axs = plt.subplots(5, 5, figsize=(25, 25))
for i in range(len(axs)):
    for j in range(len(axs[0])):
        axs[i][j].axis("off")
        axs[i][j].imshow(plt.imread(random.choice(image_paths)))
```

```python
img = Image.new("RGB", (256, 256), "#ffffff")
draw = ImageDraw.Draw(img)
r1, c1 = randint(0, 200), randint(0, 200)
r2, c2 = randint(r1, 256), randint(c1, 256)
draw.line((r1, c1, r2, c2), fill=0, width=5)
```
